```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_style()
```

```python
pdb_data_no_dups = pd.read_csv('pdb_data_no_dups.csv')
pdb_data_no_dups.head()
```

#### Removing unrelevent columns

```python
pdb_data_no_dups.drop([ 'pdbxDetails' , 'crystallizationTempK' , 'crystallizationMethod' , 'publicationYear'], axis=1 ,inplace=True)
```

### All the features

```python
for feature in pdb_data_no_dups:
    print(feature)
```

#### Filling all null values to 0

```python
pdb_data_no_dups.fillna(0 , inplace=True)
```

### All the classes of protein available

```python
print("class , count")
for index , value in pdb_data_no_dups["classification"].value_counts().items() :
    print(f"{index} , {value}")
```

### Experimental Techniques

```python
print("experimentalTechnique , count")
for index, value in pdb_data_no_dups['experimentalTechnique'].value_counts().items():
    print(f"{index} , {value}")
```

### Macromolecule Type

```python
print("macromoleculeType , count")
for index , value in pdb_data_no_dups["macromoleculeType"].value_counts().items() :
    print(f"{index} , {value}")
```

```python
pdb_data_seq = pd.read_csv('pdb_data_seq.csv')
pdb_data_seq.head()
```

```python
for feature in pdb_data_seq:
    print(feature)
```

```python

common_columns = pdb_data_no_dups.columns.intersection(pdb_data_seq.columns).tolist()

protein_dataset = pd.merge(pdb_data_no_dups, pdb_data_seq, on=common_columns, how='right')
protein_dataset.head()
```
